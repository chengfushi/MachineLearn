```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import  seaborn as sns
data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep=r"\s+", skiprows=22, header=None)
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]
```

```python
df = pd.DataFrame(data, columns=['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD',
       'TAX', 'PTRATIO', 'B', 'LSTAT'])
df['price'] = target
df.head()
```

```python
def find_price_by_similar(rm_to_price,query_x,topn = 3):
    # key=lambda x_y:(x_y[0] - query_x)**2 计算将房价的差值的平方，将其作为排序的依据
    # sorted(rm_to_price.items(),key=lambda x_y:(x_y[0] - query_x)**2) 按照差值的平方进行排序
    # [:topn] 取前topn个元素
    # np.mean([p for x,p in sorted(rm_to_price.items(),key=lambda x_y:(x_y[0] - query_x)**2)][:topn]) 计算前topn个元素的平均值
    return np.mean([p for x,p in sorted(rm_to_price.items(),key=lambda x_y:(x_y[0] - query_x)**2)][:topn])
```
